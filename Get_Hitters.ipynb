```python
# pip install pandas requests
import requests, pandas as pd
from typing import Dict, Any, List

BASE = "https://statsapi.mlb.com/api/v1/stats"

def _fetch_stats(params: Dict[str, Any]) -> List[Dict[str, Any]]:
    # Ensure we’re not page-limited
    base_params = dict(limit=10000, playerPool="ALL")  # key fix
    q = {**base_params, **params}
    try:
        r = requests.get(BASE, params=q, timeout=30)
        r.raise_for_status()
        data = r.json() or {}
        stats = data.get("stats", [])
        return (stats[0].get("splits", []) if stats else []) or []
    except Exception:
        return []

def _extract_df(splits: List[Dict[str, Any]], tag: str) -> pd.DataFrame:
    rows = []
    for s in splits:
        player = s.get("player") or s.get("person") or {}
        team   = s.get("team", {}) or {}
        stat   = s.get("stat", {}) or {}
        rows.append({
            "player_id":   player.get("id"),
            "player_name": player.get("fullName"),
            "team_id":     team.get("id"),
            "team_name":   team.get("name"),
            f"AB_{tag}":   stat.get("atBats", 0),
            f"H_{tag}":    stat.get("hits", 0),
            f"2B_{tag}":   stat.get("doubles", 0),
            f"3B_{tag}":   stat.get("triples", 0),
            f"HR_{tag}":   stat.get("homeRuns", 0),
            f"SF_{tag}":   stat.get("sacFlies", 0),
            f"BB_{tag}":   stat.get("baseOnBalls", 0),
            f"HBP_{tag}":  stat.get("hitByPitch", 0),
            f"PA_{tag}":   stat.get("plateAppearances", 0),
        })
    df = pd.DataFrame(rows)
    if df.empty:
        return df
    df = df.dropna(subset=["player_id"]).astype({"player_id": int})
    for c in df.columns:
        if c.endswith(f"_{tag}") and c not in ("player_id",):
            df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)
    return df

def _consolidate_players(df: pd.DataFrame, tag: str) -> pd.DataFrame:
    """Sum counting stats across team rows per player; keep a representative name/team."""
    if df.empty: return df
    agg_cols = [c for c in df.columns if c.endswith(f"_{tag}")]
    # pick the team row with the most AB for display
    meta = df.sort_values(f"AB_{tag}", ascending=False).drop_duplicates("player_id")[["player_id","player_name","team_name"]]
    totals = df.groupby("player_id", as_index=False)[agg_cols].sum()
    out = meta.merge(totals, on="player_id", how="left")
    return out

def fetch_season_hitting(season: int, tag: str) -> pd.DataFrame:
    params = dict(stats="season", group="hitting", season=season, gameType="R", sportId=1)
    return _consolidate_players(_extract_df(_fetch_stats(params), tag), tag)

def fetch_range_hitting(start_mmddyyyy: str, end_mmddyyyy: str, season: int, tag: str) -> pd.DataFrame:
    # MLB examples use MM/DD/YYYY for byDateRange
    params = dict(stats="byDateRange", group="hitting", gameType="R",
                  sportId=1, season=season, startDate=start_mmddyyyy, endDate=end_mmddyyyy)
    return _consolidate_players(_extract_df(_fetch_stats(params), tag), tag)

def compute_slg(df: pd.DataFrame, tag: str) -> pd.Series:
    AB  = pd.to_numeric(df[f"AB_{tag}"], errors="coerce").fillna(0)
    H   = pd.to_numeric(df[f"H_{tag}"], errors="coerce").fillna(0)
    _2B = pd.to_numeric(df[f"2B_{tag}"], errors="coerce").fillna(0)
    _3B = pd.to_numeric(df[f"3B_{tag}"], errors="coerce").fillna(0)
    HR  = pd.to_numeric(df[f"HR_{tag}"], errors="coerce").fillna(0)
    _1B = (H - _2B - _3B - HR).clip(lower=0)
    TB  = _1B + 2*_2B + 3*_3B + 4*HR
    slg = pd.Series(0.0, index=df.index)
    nz = AB > 0
    slg.loc[nz] = (TB[nz] / AB[nz]).astype(float)
    return slg

# ... all the same imports and functions as before ...

def main(top_n: int = 130):
    s25   = fetch_season_hitting(2025, "s25")
    s24   = fetch_season_hitting(2024, "s24")
    jul25 = fetch_range_hitting("07/01/2025", "07/31/2025", 2025, "jul25")

    # quick sanity
    print(f"s25 rows: {len(s25):>4}  | unique players: {s25['player_id'].nunique():>4}")
    print(f"s24 rows: {len(s24):>4}  | unique players: {s24['player_id'].nunique():>4}")
    print(f"jul25 rows: {len(jul25):>4}| unique players: {jul25['player_id'].nunique():>4}")

    base = s25.merge(s24[["player_id","AB_s24"]], on="player_id", how="left") \
              .merge(jul25[["player_id","AB_jul25"]], on="player_id", how="left")

    base["AB_s24"]   = pd.to_numeric(base["AB_s24"], errors="coerce").fillna(0).astype(int)
    base["AB_jul25"] = pd.to_numeric(base["AB_jul25"], errors="coerce").fillna(0).astype(int)

    # Thresholds
    mask = (base["AB_s25"] >= 350) & (base["AB_s24"] >= 350) & (base["AB_jul25"] >= 80)
    qualified_ids = set(base.loc[mask, "player_id"])

    q = s25[s25["player_id"].isin(qualified_ids)].copy()
    if q.empty:
        print("No players matched the thresholds — check the counts printed above.")
        return

    q["SLG_2025"] = compute_slg(q, "s25")
    q = q.merge(s24[["player_id","AB_s24"]], on="player_id", how="left") \
         .merge(jul25[["player_id","AB_jul25"]], on="player_id", how="left")
    q["AB_s24"]   = q["AB_s24"].fillna(0).astype(int)
    q["AB_jul25"] = q["AB_jul25"].fillna(0).astype(int)

    out_cols = ["player_name","team_name","AB_s25","AB_s24","AB_jul25","SLG_2025"]
    # 👇 flip sort order so lowest SLG is on top
    q = q[out_cols].sort_values("SLG_2025", ascending=True).reset_index(drop=True)

    print(f"\nQualified hitters: {len(q)}")
    pd.set_option("display.max_rows", 200)
    print(q.head(top_n).to_string(index=False))

if __name__ == "__main__":
    main(top_n=130)
```
